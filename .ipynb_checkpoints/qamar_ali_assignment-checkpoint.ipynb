## Customer Segmentation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler


# Merge data for feature creation
merged_transactions = tran.merge(prod, on="ProductID", how="left")

# Calculate total spending per product category for each customer
category_spending = merged_transactions.pivot_table(
    index="CustomerID", columns="Category", values="TotalValue", aggfunc="sum", fill_value=0
)
category_spending.columns = [f"Spending_{col}" for col in category_spending.columns]

# Aggregate transaction data for clustering
transactions_aggregated = tran.groupby("CustomerID").agg({
    "TotalValue": ["sum", "mean", "count"] 
}).reset_index()
transactions_aggregated.columns = ["CustomerID", "TotalSpending", "AvgTransactionValue", "TransactionFrequency"]

# Merge customer, transaction, and category spending data
customer_features = cust.merge(transactions_aggregated, on="CustomerID", how="left").merge(
    category_spending, on="CustomerID", how="left"
).fillna(0)

#Normalize numerical features
scaler = MinMaxScaler()
numeric_features = ["TotalSpending", "AvgTransactionValue", "TransactionFrequency"] + list(category_spending.columns)
customer_features[numeric_features] = scaler.fit_transform(customer_features[numeric_features])

# Perform clustering(2 to 10)
clustering_features = customer_features[numeric_features]
dbi_scores = {}
cluster_results = {}

for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(clustering_features)
    dbi = davies_bouldin_score(clustering_features, cluster_labels)
    dbi_scores[n_clusters] = dbi
    cluster_results[n_clusters] = cluster_labels

# Find the optimal number of clusters based on the lowest DBI score
optimal_clusters = min(dbi_scores, key=dbi_scores.get)

# Add the cluster labels for the optimal number of clusters to the data
customer_features["Cluster"] = cluster_results[optimal_clusters]

# Visualize the clustering results using PCA
pca = PCA(n_components=2)
pca_features = pca.fit_transform(clustering_features)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=pca_features[:, 0],
    y=pca_features[:, 1],
    hue=customer_features["Cluster"],
    palette="tab10",
    alpha=0.8
)
plt.title(f"Customer Segmentation with {optimal_clusters} Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()

#Output DBI scores for all cluster configurations
print("Davies-Bouldin Index Scores:")
for n_clusters, dbi in dbi_scores.items():
    print(f"Clusters: {n_clusters}, DBI: {dbi:.4f}")

#Save the clustered data to a CSV file
customer_features.to_csv("Customer_Segmentation_Clusters.csv", index=False)
print("Clustered data saved to 'Customer_Segmentation_Clusters.csv'")
```
